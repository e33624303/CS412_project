```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import normalize
from sklearn.cluster import DBSCAN
from data import data_init
import matplotlib.pyplot as plt
```

```python
data = data_init()
```

```python
X = np.array([data['Population'],data['Population']*data['Population']]).T.reshape(-1, 2)
Y = np.array(data['Air Pollution']).T
regr = LinearRegression(normalize=True).fit(X, Y)
Y_pred = regr.predict(X)
loss = np.mean((Y-Y_pred)*(Y-Y_pred))
print('MSE = ', loss)
Xp = np.linspace(np.min(data['Population']), np.max(data['Population']), 1000)
Xp_ = np.array([Xp, Xp*Xp]).T.reshape(-1, 2)
Yp = regr.predict(Xp_)
plt.plot(Xp, Yp)
plt.scatter(data['Population'], Y,c='r')
plt.xlabel("Population")
plt.ylabel("Air Pollution")
plt.show()
```
